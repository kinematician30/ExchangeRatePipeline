```python
import requests
import pandas
import json
```

```python
date = '2025-03-29'
base_url = f'https://{date}.currency-api.pages.dev/v1/currencies/ngn.json'

resp = requests.get(base_url)

data = resp.json()
```

```python
dt = data['date']
rates = data['ngn']
for k, v in rates.items():
    print(f"{dt} - {k} - {v:.4f}")
```
